```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the retail sales dataset
dataset_path = r"C:\Users\ARYAN PARIKH\Downloads\archive\retail_sales_dataset.csv"
retail_data = pd.read_csv(dataset_path)

# Display the first few rows of the dataset
print(retail_data.head())

# Data Cleaning (if needed)
# Example: Remove missing values
# retail_data.dropna(inplace=True)

# Check the data type of the 'Total Amount' column
print(retail_data['Total Amount'].dtype)

# Convert 'Total Amount' column to numeric (if needed)
retail_data['Total Amount'] = pd.to_numeric(retail_data['Total Amount'], errors='coerce')

# Descriptive Statistics
descriptive_stats = retail_data.describe()
print(descriptive_stats)

# Time Series Analysis
# Example: Convert the 'Date' column to datetime
retail_data['Date'] = pd.to_datetime(retail_data['Date'])
# Example: Plot sales trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Total Amount', data=retail_data)
plt.title('Retail Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.show()

# Customer and Product Analysis
# Example: Customer demographics
customer_demographics = retail_data['Gender'].value_counts()
print(customer_demographics)

# Example: Product analysis
product_analysis = retail_data.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)
print(product_analysis)

# Visualization
# Example: Bar chart for Product Category Sales
plt.figure(figsize=(10, 6))
sns.barplot(x='Total Amount', y='Product Category', data=retail_data, estimator=sum, ci=None)
plt.title('Total Sales by Product Category')
plt.xlabel('Total Amount')
plt.ylabel('Product Category')
plt.show()

# Recommendations
# Provide actionable recommendations based on insights gained from the EDA

# Save the cleaned dataset (if needed)
retail_data.to_csv("cleaned_retail_data.csv", index=False)
```
